```python
import numpy as np
```

```python
row1 = np.array([40,45,25,10])
row2 = np.array([35,30,20,30])
```

```python
np.sum(row1) + np.sum(row2)
```

```python
sum_r1 = np.sum(row1)
sum_r2 = np.sum(row2)
```

```python
sum_row=np.array([sum_r1,sum_r2])
```

```python
sum_col = row1+row2
```

```python
sum_col
```

```python
sum_row
```

```python
exp = []
for i in sum_row:
    for j in sum_col:
        value = (i*j)/235
        exp.append(value)
```

```python
exp
```

```python
obj = np.array([40,45,25,10,35,30,20,30])
```

```python
obj
```

```python
np.sum((np.square(obj-exp)/exp))
```
