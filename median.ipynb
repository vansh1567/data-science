```python
from statistics import median, mean
from unittest.mock import inplace

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/titanic.csv")
```

```python
df.head()
```

```python
np.median(df['Age'])
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
age_mn = np.mean(df['Age'])
```

```python
age_mn
```

```python
age_md = df['Age'].median()
```

```python
age_md
```

```python
sns.histplot(x='Age', data=df, bins=range(0,81,10))
plt.plot([age_mn for i in range(0,300)],[i for i in range(0,300)], color='red')
plt.plot([age_md for i in range(0,300)],[i for i in range(0,300)], color='blue')
plt.show()
```
