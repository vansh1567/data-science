if X is incresing and Y too its a positive covariance

if X decreasing and Y is increasing then its a negative covariance

if x increased and there is no change in Y thats mean no covariance

Range of covariance is -infinity to infinity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/tips.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
data_corr = df.select_dtypes('float64','int64').corr()
data_corr
```

```python
data_cov = df.select_dtypes('float64','int64').cov()
data_cov
```

```python
sns.heatmap(data_corr,annot=True)
plt.show()
```

```python
sns.heatmap(data_cov,annot=True)
plt.show()
```
