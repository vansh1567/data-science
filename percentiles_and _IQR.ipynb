Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/titanic.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# df['Age'].fillna(df['Age'].mean() , inplace=True) this used to work in older version of pandas
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
df['Age']
```

```python
# np.percentile(df['Age'], [25, 50, 75])
np.percentile(df["Age"],25)
```

```python
np.percentile(df['Age'],50) , np.percentile(df['Age'],75), np.percentile(df['Age'],100)
```

```python
np.percentile(df['Age'],0)
```

```python
sns.boxplot(x="Age", data=df)
plt.show()
```

```python
sns.boxplot(x="Fare",data=df)
plt.show()
```
